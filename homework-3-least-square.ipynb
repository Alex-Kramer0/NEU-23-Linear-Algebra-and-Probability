In this question, you will solve least squares through projection using python. Please find the attached dataset “Real estate.csv”. You should follow the projection approach from section 4.3. Plot your predictions and report your results.
*   Predict Y (price) using house age.
*   Predict Y using another feature.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# reading data
data = pd.read_csv('Real estate.csv')
print(data.shape)
print(data.head())

# x = house age
# y = price 
x = data['X2 house age'].values
y = data['Y house price of unit area'].values

# create matrix A
a = ''
for i in range(0, len(x)):
  if i == 0:
    a += '1 ' + str(x[i])
  else:
    a += '; ' + '1 ' + str(x[i])
  # creates a

matrix_a = np.matrix(a)

# create matrix b -- predicted price -- 1D array
b = ''
for e in range(0, len(y)):
  if e == 0:
    b += str(y[e])
  else:
    b += '; ' + str(y[e])

matrix_b = np.matrix(b)

# get A^t (A transpose)
a_transpose = matrix_a.transpose()

# multiply A^t*A (A Transpose times A)
ata = np.matmul(a_transpose, matrix_a)

# multiply A^t*B (A transpose times matrix b)
atb = np.matmul(a_transpose, matrix_b)

# get inverse of ata 
i_ata = ata.getI()

# multply inverse of ata by atb
x_hat = np.matmul(i_ata, atb)

c = float(x_hat[0]) # y-intercept

d = float(x_hat[1]) # slope
print('Coefficents: ')
print('C: ', c, 'D: ', d)
# plot values and regression line
max_x = np.max(x)
min_x = np.min(x)

# calculate line values x and y 
x_value = np.linspace(min_x, max_x, 1000)
y_value = c + d * x_value

# plot line
plt.plot(x_value, y_value, color='#58b970', label='Regression Line')

# plot scatter points
plt.scatter(x, y, c='#ef5423', label='Scatter Plot')

plt.xlabel('House Age')
plt.ylabel('House Price of Unit Area')
plt.legend()
plt.show()
```
